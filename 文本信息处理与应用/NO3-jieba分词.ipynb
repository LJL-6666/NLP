# 数据导入

```python
import pandas as pd
#1、读入为数据框
#有的环境配置下，read_table出错，改用read_csv
raw=pd.read_table('./PythonData/金庸-射雕英雄传txt精校版.txt',names=['txt'],encoding='GBK')
```

```python
print(len(raw))
raw.head()
```

```python
# 2、加入章节标识符
#章节判断用变量预处理
def m_head(tmpstr):
    return tmpstr[:1]
def m_mid(tmpstr):
    return tmpstr.find("回 ")
raw['head']=raw.txt
```

```python
raw['head']=raw.txt.apply(m_head)
raw['mid']=raw.txt.apply(m_mid)
raw['len']=raw.txt.apply(len)
raw.head(50)
```

```python
#3、章节判断
chapnum=0
for i in range(len(raw)):
    if raw['head'][i]=='第' and raw['mid'][i]>0 and raw['len'][i]<30:
        chapnum+=1
    if chapnum>=40 and raw['txt'][i]=='附录一：成吉思汗家族':
        chapnum=0
    raw.loc[i,'chap']=chapnum
```

```python
del raw['head']
del raw['mid']
del raw['len']
```

```python
raw.head()
```

```python
# 3、提取出所需章节
```

```python
raw[raw.chap==1]
```

```python
chapter=raw.groupby('chap').agg('sum')
```

```python
chapter=chapter[chapter.index!=0]
```

```python
chapter.index=chapter.index.astype('int')
```

```python
chapter_1=chapter.txt[1]
```
